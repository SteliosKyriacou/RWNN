```python
import tensorflow.keras as keras
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from RWNN import PseudoRWNN
from RWNN import EnsemblePsudoRWNN
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
#1D
#simple 1d input -> 1d output training set (easy visualization)
seed=0
rng=np.random.RandomState(seed)
x=rng.rand(1000,1)
y=(x-0.5)**2
plt.plot(x,y,'.',)
```

```python
# Build a 3 layer Pseudo RWNN and train it 
layers=[
    (200,'sigmoid',.5),
    (40,'sigmoid',.5),
    (1,'linear',1) # 1d output 
]

ens = EnsemblePsudoRWNN.EnsemblePsudoRWNN(3,1,layers,seed=42)
ens.train(x,y,batch_size=16,epochs=160,learning_rate=0.2)
```

```python
yy=ens.use(x)

for yi in yy:
    plt.plot(x,yi,'.',c='r')
    
plt.plot(x,y,'o')
```

```python
plt.plot(x,y,'.')
plt.plot(x,np.mean(yy,axis=0),'.')
```
